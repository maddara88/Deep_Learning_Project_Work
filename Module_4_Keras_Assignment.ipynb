```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df = pd.read_csv("C:\\Users\\HP\\Downloads\\diabetes.csv")
```

```python
df.head()
```

```python
#Here we are going to built our first model with using ‘Pregnancies’, ‘Glucose’ & ‘BloodPressure’ as 
# independent columns and outcome as dependent column
```

# Model1

```python
x = df.loc[:,["Pregnancies","Pregnancies","BloodPressure"]]
```

```python
x.head()
```

```python
#Now we are going to do scaling

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled[1].shape
```

```python
y = df["Outcome"]
```

```python
y.head()
```

```python
y_cat = tf.keras.utils.to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y_cat,test_size = 0.1,random_state = 47)
```

```python
#del model1
model1 = tf.keras.models.Sequential()
model1.add(tf.keras.Input((3,),name = "input"))
# Add the flatten layers to the model
model1.add(tf.keras.layers.Flatten(name = "Flat"))
model1.add(tf.keras.layers.Dense(8,activation = "relu",name = "Dense1"))
model1.add(tf.keras.layers.Dense(2,activation = "softmax",name = "OutputLayer"))
```

```python
model1.compile(optimizer = tf.keras.optimizers.SGD(learning_rate = 0.01),loss = tf.keras.losses.categorical_crossentropy,
             metrics = ["acc"])
```

```python
model1_history = model1.fit(x_train,y_train,epochs = 100,validation_split = 0.1,batch_size = 10)
```

```python
#Here in our model1 by taking ‘Pregnancies’, ‘Glucose’ & ‘BloodPressure’ as 
# independent columns and outcome as dependent column we got an val_acc score as 64% and train_acc as 67% with a 
#val loss 64
```

# Model2

```python
# B. Build another sequential model where ‘Outcome’ is the dependent variable and all 
# other columns are predictors. 
# a. This model should have 3 hidden layers with 16 nodes in each layer 
# b. Use ‘adam’ as the optimization algorithm 
# c. Fit the model, with number of epochs to be 150 and batch size to be 10
```

```python
x2 = df.iloc[:,:-1]
```

```python
x2.head()
```

```python
y.head()
```

```python
x2_scaled = scaler.fit_transform(x2)
y2_cat = tf.keras.utils.to_categorical(y)
```

```python
x2_scaled[1].shape
```

```python
x2_train,x2_test,y2_train,y2_test = train_test_split(x2_scaled,y2_cat,test_size = 0.1,random_state = 47)
```

```python
#del model2
model2 = tf.keras.models.Sequential()
model2.add(tf.keras.Input((8,),name = "input"))
# Adding the flatten layers to the model
model2.add(tf.keras.layers.Flatten(name = "Flat"))
model2.add(tf.keras.layers.Dense(16,activation = "relu",name = "Dense1"))
model2.add(tf.keras.layers.Dense(16,activation = "relu",name = "Dense2"))
model2.add(tf.keras.layers.Dense(16,activation = "relu",name = "Dense3"))
model2.add(tf.keras.layers.Dense(2,activation = "softmax",name = "OutputLayer"))
```

```python
model2.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),loss = tf.keras.losses.categorical_crossentropy,
             metrics = ["acc"])
```

```python
model2_history = model2.fit(x2_train,y2_train,epochs = 150,validation_split = 0.1,batch_size = 10)
```

```python
#Here in our model12 by taking all the fetaures  as 
# independent columns and outcome as dependent column we got an val_acc score as 84% and train_acc as 79% with a 
#val loss 45 which way better that the first model1 so thus we can say that by incresing the number of nurons and 
# by applying adam optimizer instead of SGD we got a better accuracy score
```
